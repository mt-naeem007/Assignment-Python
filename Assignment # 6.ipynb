```python
class car():
    #instantiation method
    def __init__(self,companyname,category,year,color,price): 
    
    
        self.companyname =companyname
        self.category =category
        self.year = year
        self.color = color
        self.price = price
        
        
   
    def updateprice(self,newprice):   
            self.newprice = newprice

    def updatecolor(self,newcolor):   
            self.newcolor = newcolor
    
    def updateversion(self,newcategory):   
            self.newcategory = newcategory        
             

    
auto1=car('toyota corolla', 'S class', '2010' ,'white colour', ' RS 1200000')       
print(auto1.companyname)
print(auto1.category)
print(auto1.year)
print(auto1.color)
print(auto1.price)

auto1.updateprice(' RS 1500000')
print(auto1.newprice," updated price")

auto1.updatecolor('green colour')
print(auto1.newcolor)

auto1.updateversion('A class ')
print(auto1.newcategory)
print("\n")



auto2=car('suzuki Mehran', 'S class', '2005' ,'blue colour', ' RS 400000')       
print(auto2.companyname)
print(auto2.category)
print(auto2.year)
print(auto2.color)
print(auto2.price)

auto2.updateprice(' RS 800000')
print(auto2.newprice," updated price")

auto2.updatecolor('red colour')
print(auto2.newcolor)

auto2.updateversion('C class ')
print(auto2.newcategory)
print("\n")



auto2=car('Honda Civic', 'S class', '2005' ,'silver colour', ' RS 1000000')       
print(auto2.companyname)
print(auto2.category)
print(auto2.year)
print(auto2.color)
print(auto2.price)

auto2.updateprice(' RS 1200000')
print(auto2.newprice," updated price")

auto2.updatecolor('red colour')
print(auto2.newcolor)

auto2.updateversion('A class ')
print(auto2.newcategory)
print("\n")



auto2=car('suzuki Cultus', 'S class', '2010' ,'white colour', ' RS 400000')       
print(auto2.companyname)
print(auto2.category)
print(auto2.year)
print(auto2.color)
print(auto2.price)

auto2.updateprice(' RS 800000')
print(auto2.newprice," updated price")

auto2.updatecolor('golden colour')
print(auto2.newcolor)

auto2.updateversion('C class ')
print(auto2.newcategory)
print("\n")




auto2=car('Toyota vigo champ', 'S class', '2012' ,'white colour', ' RS 4000000')       
print(auto2.companyname)
print(auto2.category)
print(auto2.year)
print(auto2.color)
print(auto2.price)

auto2.updateprice(' RS 5000000')
print(auto2.newprice," updated price")

auto2.updatecolor('black colour')
print(auto2.newcolor)

auto2.updateversion('A class ')
print(auto2.newcategory)
print("\n")



```
